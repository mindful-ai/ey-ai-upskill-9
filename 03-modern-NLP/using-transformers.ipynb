### Huggingface - Basic Inference

```python
from transformers import pipeline
```

```python
model = pipeline("sentiment-analysis")
model
```

```python
model("Today is a great day for learning AI")
```

```python
model("I am not sure if it is a good day for transformers")
```

### Classification

```python
model = pipeline('zero-shot-classification', candidate_labels=["sports", "politics", "business", "education"])
```

```python
model("This is a course on Huggingface")
```

```python
model("The stock prices for Tesla is plummating")
```

### Text Generation

```python
model = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M-Instruct")
```

```python
text = model("Complete the statement, give a paragraph of 100 words: gravity is a force that ...")
```

```python
print(text[0]["generated_text"])
```

### Wrong Use Case

```python
model('Classify the statement into ["sports", "politics", "business", "education"]: This is a course on Huggingface')
```
